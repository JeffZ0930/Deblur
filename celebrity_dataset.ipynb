## Test Gaussian Blur


```python
import os
import numpy as np
import cv2
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from torchvision.utils import save_image
```

```python
female_origin = "./female_origin"
male_origin = "./male_origin"
female_blur = "./female_blur"
male_blur = "./male_blur"

level_1 = 15
level_2 = 25
level_3 = 45

# FeMale
for img_name in os.listdir(female_origin):
    img_path = os.path.join(female_origin, img_name)
    img = cv2.imread(img_path)

    blurred_1 = cv2.GaussianBlur(img, (level_1, level_1), 0)
    output_path = os.path.join(female_blur, "level_1/", img_name)
    cv2.imwrite(output_path, blurred_1)

    blurred_2 = cv2.GaussianBlur(img, (level_2, level_2), 0)
    output_path = os.path.join(female_blur, "level_2/", img_name)
    cv2.imwrite(output_path, blurred_2)

    blurred_3 = cv2.GaussianBlur(img, (level_3, level_3), 0)
    output_path = os.path.join(female_blur, "level_3/", img_name)
    cv2.imwrite(output_path, blurred_3)

# Male
for img_name in os.listdir(male_origin):
    img_path = os.path.join(male_origin, img_name)
    img = cv2.imread(img_path)

    blurred_1 = cv2.GaussianBlur(img, (level_1, level_1), 0)
    output_path = os.path.join(male_blur, "level_1/", img_name)
    cv2.imwrite(output_path, blurred_1)

    blurred_2 = cv2.GaussianBlur(img, (level_2, level_2), 0)
    output_path = os.path.join(male_blur, "level_2/", img_name)
    cv2.imwrite(output_path, blurred_2)

    blurred_3 = cv2.GaussianBlur(img, (level_3, level_3), 0)
    output_path = os.path.join(male_blur, "level_3/", img_name)
    cv2.imwrite(output_path, blurred_3)
```

```python
clear_transforms = transforms.Compose([
    # transforms.RandomResizedCrop(224),
    # transforms.RandomHorizontalFlip(0.1),
    # transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    # transforms.RandomRotation(20),
    # transforms.RandomVerticalFlip(0.1),
    transforms.ToTensor(),
    # transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Assuming ImageNet mean and std
])

dataset = ImageFolder(root='male_origin', transform=clear_transforms)

train_size = int(0.7 * len(dataset))  # 70% for training
val_size = int(0.15 * len(dataset))   # 15% for validation
test_size = len(dataset) - train_size - val_size  # Remaining for testing


train_size = int(0.7 * len(dataset))  # 70% for training
val_size = int(0.15 * len(dataset))   # 15% for validation
test_size = len(dataset) - train_size - val_size  # Remaining for testing

train_dir = "datasets/clear/train"
val_dir = "datasets/clear/val"
test_dir = "datasets/clear/test"

train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

for dataset, output_folder in [(train_dataset, train_dir), (val_dataset, val_dir), (test_dataset, test_dir)]:
    for idx in range(len(dataset)):
        image, _ = dataset[idx]
        output_path = os.path.join(output_folder, f"{idx}.jpg")
        save_image(image, output_path)
```
